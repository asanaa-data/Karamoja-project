#### Renaming the column names in the district and sub-county datasets
----------------------------------------------------------------

```python
# renaming the columns in the district dataset
dt.columns = ['District', 'Population_dt', 'Area_dt', 'Sorghum_Yield_dt (Ha)', 'Maize_Yield_dt (Ha)', 'Crop_Area_dt (Ha)',
       'Sorghum_Area_dt (Ha)', 'Maize_Area_dt (Ha)', 'Sorghum_Production_Total_dt', 'Maize_Production_Total_dt']
# View if the changes took place
dt.columns
```

#### Another way to rename columns is by using the 'rename() function. The syntax would be
- dt.rename(columns = {'NAME':'District_name'})
- dt.rename(columns = {'POP':'Population'}) etc

```python
# renaming the columns in the sub-county dataset
st.columns = ['Sub-county', 'District', 'Population_St', 'Area_st', 'Karamoja',
       'Sorghum_Yield_st (Ha)', 'Maize_Yield_st (Ha)', 'Crop_Area_st (Ha)', 'Sorghum_Area_st (Ha)', 'Maize_Area_st (Ha)',
       'Sorghum_Production_Total_st', 'Maize_Production_Total_st']
# View if the changes took place
st.columns
```

#### Changing the case of the district column and sub-county names column (from UPPER case to Title case)

```python
# using the lambda function to conver the values under the districts' column from UPPER case to Sentence/Title case
# i shall creat a column for this task
dt['District'] = dt['District'].map(lambda x : x.title())
dt.head(5)
```

```python
st['District'] = st['District'].map(lambda x : x.title())
st['Sub-county'] = st['Sub-county'].map(lambda x : x.title())
st.head(5)
```

#### Deleting the column named 'Karamoja' in the sub-county dataset because it does not seem to mean anything or affect the rest of the columns

```python
# First checking the number of (rows, columns) in the sub-county dataset before deleting the columns mentioned
st.shape
```

```python
# deleting the 'Karamoja' column in the sub-county dataset
del(st['Karamoja'])
```

### 5. Exploring Data with visualiations

##### Population distribution in Karamoja sub-region

```python
# plotting a histogram for to show the distribution of the Karamoja region population
plt.hist(x=dt['Population_dt'],edgecolor='Black', bins=14, color='y')
plt.title('District population distribution')
plt.xlabel('Population')
plt.ylabel('count')

# calculating the mean, median and mode using pandas

# the mean/average population column using the pandas mean() function
mean_Population_dt = dt['Population_dt'].mean()
# the median of population column using the pandas median() function
median_Population_dt = dt['Population_dt'].median()
# the mode of population column using the pandas mode() function
mode_Population_dt = dt['Population_dt'].mode()
# the standard deviation of population column using the pandas std() function
stdDev_Population_dt = dt['Population_dt'].std()

# with the help of the axvline() function we can plot inside the histogram/ Add a vertical line across the Axes
plt.axvline(mean_Population_dt, color='black', linestyle='-')
#plt.axvline(median_population_dt, color='y', linestyle='-')
#plt.axvline(mode_population_dt, color='b', linestyle='-')
#plt.legend()

plt.show()
```

##### Maize & Sorghum yield correlation with Maize & Sorghum area (with seaborn)

```python

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,7))

# Plot distribution of sorghum yield
sns.scatterplot(data=st, x=st['Sorghum_Area_st (Ha)'], y=st['Sorghum_Yield_st (Ha)'],ax=ax1, 
                hue=st['Sorghum_Production_Total_st'], 
                size=st['Sorghum_Production_Total_st'],
               palette="gist_heat_r")
# Plot distribution of maize yield
sns.scatterplot(data=st, x=st['Maize_Area_st (Ha)'], y=st['Maize_Yield_st (Ha)'],ax=ax2, 
                hue=st['Maize_Production_Total_st'], 
                size=st['Maize_Production_Total_st'],
                palette="Blues_r" )

# Customize labels
ax1.set_title('Distribution of correlation between Sorghum area & Sorghum yield')
ax1.set_xlabel('Sorghum Area')
ax1.set_ylabel("Sorghum yield")
ax2.set_title('Distribution of correlation between Maize area & Maize yield')
ax2.set_xlabel('Maize area')
ax2.set_ylabel("Maize yield");
```

##### Maize & Sorghum yield correlation with Maize & Sorghum area (with pandas)

```python
#define subplot layout
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(17, 5)) # the nrows & ncols spacify the layout of our subplots


#add DataFrames to subplots for boxplots
# axes[0].hist(dt['Sorghum_Yield_dt (Ha)'], showfliers=False) # the showfliers=False helps hide the outliers in the graph
axes[0].boxplot(dt['Sorghum_Yield_dt (Ha)'])
axes[1].hist(dt['Population_dt'],edgecolor='black', bins=13, color='green')
axes[2].hist(dt['Area_dt'],edgecolor='black', bins=10)

# boxplot labels

# labels for the Sorghum yield plot
axes[0].set_title("Distributions of Sorghum yield in the region")
axes[0].set_ylabel("Sorghum yield")
# labels for the population plot
axes[1].set_title("Distributions of population in the region")
axes[1].set_ylabel("Population")
# labels for the district area plot
axes[2].set_title("District area")
axes[2].set_ylabel("size")
plt.show()

print('Subcounty plots')


fig1, axes1 = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))
#add DataFrames to subplots for plot
axes1[0].boxplot(st['Maize_Yield_st (Ha)'])
axes1[1].hist(st['Population_St'],edgecolor='black', bins=13)
axes1[2].scatter(st['Area_st'],st['Maize_Production_Total_st'],color='navy')

# Histogram labels

# labels for the maize yield plot
axes1[0].set_title("Distributions of Sorghum yield in the region")
axes1[0].set_xlabel("Maize yield")
# labels for the sub-county population above grade plot
axes1[1].set_title("Population distribution at subcounty")
axes1[1].set_xlabel("sub-county population")
# labels for the area in sub-county plot
axes1[2].set_title("correlation (maize yield & subcounty area)")
axes1[2].set_xlabel("maize yield")
axes1[2].set_ylabel("subcounty area")
plt.show()
```
